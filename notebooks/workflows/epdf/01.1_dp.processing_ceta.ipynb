```python
def show_image(img, title="", cmap="cividis", vmin=None, vmax=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.title(title)
    plt.axis("off")
    plt.show()

def log_scale(img):
    img = img.astype(np.float32)
    return np.log1p(img)

def normalize(img, clip_percent=0.2):
    img = img.astype(np.float32)
    low, high = np.percentile(img, [clip_percent, 100 - clip_percent])
    img = np.clip(img, low, high)
    img -= img.min()
    img /= img.max()
    return img

show_image(normalize(img), title="Percentile Contrast (1% clipped)")

#show_image(log_scale(img), title="Log Scale")

img_out = normalize(img)

from PIL import Image

Image.fromarray(img_out).save(r"C:\Users\seccolev\data_processing\data\processed\ePDF\CdSe\ED_Ilum.tif")
```

## Finding the center of the diffraction pattern

The find_center function uses the Hough circle transform and edge detection methods to locate the center of the diffraction pattern. The remainder of the cell plots the identified circle.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# --- Load image and prepare analysis ---
# img = image_array  # Uncomment if image is named image_array
analysis = ImageAnalysis()

# --- Find center using Hough transform or similar method ---
center_x, center_y, r = analysis.find_center(
    img,
    r=1,                 # Smallest radius
    R=5000,              # Largest radius
    threshold=120,       # Threshold for diffractogram
    edges_thresh1=255,   # Canny edge detector threshold 1
    edges_thresh2=10     # Canny edge detector threshold 2
)

# Create a circle patch to visualize the found center
circle1 = Circle((center_x, center_y), r, fill=False, color='white')

# --- Apply beamstopper and defect masks to the image ---
processing = ImageProcessing(directory)

#beamstop_mask = processing.load_mask('/home/ABTLUS/victor.secco/data_processing/ED/MASKS/mask_titan.tif')
#img[beamstop_mask == 255] = 0

#defects_mask = processing.load_mask('/home/ABTLUS/victor.secco/data_processing/ED/MASKS/test_defects_mask.png')
#img[defects_mask == 255] = 0

# --- Perform azimuthal integration ---
binning = 3000
data, polar_image, _ = analysis.azimuth_integration_cv2(img, center=[center_x, center_y], binning=binning)

# --- Plot results ---
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Show masked and processed image
ax[0].imshow(img, cmap='inferno')
ax[0].scatter(center_x, center_y, s=5, color='white')
ax[0].add_patch(circle1)
ax[0].axis('off')
ax[0].text(center_x + 110, center_y + 50, f'{center_x, center_y}', color='white', fontsize=13, ha='center', va='center')

# Optional zoom
# ax[0].set_xlim(1600, 3100)
# ax[0].set_ylim(2700, 1200)  # Y-axis is inverted in image coordinates

# Show radial profile
ax[1].plot(data)

# Adjust layout
plt.subplots_adjust(wspace=0, hspace=0)
plt.tight_layout(pad=0)
```
